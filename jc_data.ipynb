```python
#dependancies
import pandas as pd
import datetime as dt
import numpy as np
```

```python
#Create list of dataframes from csv files
f1 = "data/JC/JC-201901-citibike-tripdata.csv"

file_list = []
for i in range(1,10):
    file_list.append(f"data/JC/JC-20190{i}-citibike-tripdata.csv")
for i in range(0,3):
    file_list.append(f"data/JC/JC-20191{i}-citibike-tripdata.csv")
for i in range(1,10):
    file_list.append(f"data/JC/JC-20200{i}-citibike-tripdata.csv")
for i in range(0,3):
    file_list.append(f"data/JC/JC-20201{i}-citibike-tripdata.csv")
     

data_df_list = [pd.read_csv(file) for file in file_list]
```

```python
file_list
```

```python
#Create master dataframe from all individual dataframes (1 for each month of data)
data_df = data_df_list[0].append(data_df_list[1::],ignore_index=True)
data_df
```

```python
def format_date(date):
    if pd.notnull (date):
        ndate = dt.datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
        return ndate.strftime('%m/%d/%Y')
    else:
        return np.nan
    
def format_time(date):
    if pd.notnull (date):
        data = dt.datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
        return data.strftime('%H:%M')
    else:
        return np.nan
```

```python
#Create new columns using defined functions
data_df["start_date"] = data_df["starttime"].map(lambda x: format_date(str(x)))
data_df["start_time"] = data_df["starttime"].map(lambda x: format_time(str(x)))
data_df["end_date"] = data_df["stoptime"].map(lambda x: format_date(str(x)))
data_df["end_time"] = data_df["stoptime"].map(lambda x: format_time(str(x)))
```

```python
gender = {0:"Unknown",1:"Male",2:"Female"}
data_df["gender_type"] = data_df["gender"].map(gender)
```
